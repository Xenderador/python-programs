```python
with open("mcq_list.txt", "r") as file:
    x = file.read()
print(x)
```

```python
import re

a = re.split(r"\n{2,}", x)
```

```python
a
```

```python
# \d+ (.*)\nA (.*)
boobs = []
# text question 4 options answer

pattern2 = re.compile(r" ([ABCD])($|\n)")
question_pattern = re.compile(r"\d+ (.|\n)*?(?=[ABCD] )")
options_pattern = re.compile(r"([ABCD]) (.+[^\sABCD|\n])")
for text in a:
    dic = {}
    dic["text"] = text
    for match in pattern2.finditer(text):
        dic["answer"] = match.groups()[0]
    for match in question_pattern.finditer(text):
        dic["question"] = match.group()
    options = {}
    for match in options_pattern.finditer(text):
        options[match.groups()[0]] = match.groups()[1]
    dic["options"] = options
    boobs.append(dic)
```

```python
boobs
```

```python
import random


def shuffle(options):
    keys = [key for key in options.keys()]
    random.shuffle(keys)
    return keys
```

```python
print(boobs[69]["question"])
print(boobs[69]["options"])
keys = shuffle(boobs[69]["options"])
print(keys)

for key in keys:
    print(boobs[69]["options"][key])
```
